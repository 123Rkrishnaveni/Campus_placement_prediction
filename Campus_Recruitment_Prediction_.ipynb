## Checking for Outliers and Removing them

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(df['salary'])
plt.show()
```

It shows that only 'salary' column contain outliers,we removed it by limiting the values in the 'salary' column to a specified range.

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3-Q1
```

```python
lower_bound = Q1-1.5*IQR
print(lower_bound)
```

```python
upper_bound = Q3+1.5*IQR
print(upper_bound)
```

```python
df['salary'] = df['salary'].clip(lower=lower_bound, upper=upper_bound)
```

```python
sns.boxplot(df['salary'])
plt.show()
```

### Data Manipulation

```python
x = df.iloc[:,:-2]
y = df['status']
```

```python
x_obj_col = x.select_dtypes(include='O').columns
x_obj_col
```

### Handling Categorical Features

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
from sklearn.preprocessing import LabelEncoder

x[x_obj_col] = x[x_obj_col].apply(LabelEncoder().fit_transform)
x.head()
```

```python
x.dtypes
```

### Modeling and Accuracy Testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

models = [LogisticRegression(), GaussianNB(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]
```

```python
for model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model} :- \t Accuracy = {accuracy:.2f}")
```

## The Logistic Regression model is giving the accuracy of 88%
